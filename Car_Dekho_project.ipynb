```python
import pandas as pd
import numpy as np
```

```python
c=pd.read_csv('car_data.csv')
c.head()
```

```python
print(c['Seller_Type'].unique())
print(c['Fuel_Type'].unique())
print(c['Transmission'].unique())
print(c['Owner'].unique())
```

```python
c.drop(columns='Car_Name',inplace=True)
```

```python
c.head()
```

```python
c['No_of_Year']=2020-c['Year']
```

```python
c.head()
```

```python
c.drop(columns='Year',inplace=True)
```

```python
c.head()
```

```python
c=pd.get_dummies(c,drop_first=True)#drop_first =True to avoid dummy variable trap by dropping first categorical feature in each column
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.pairplot(c)
```

```python
c.head()
```

```python
X=c.iloc[:,1:]
Y=c.iloc[:,0]
```

```python
X.head()
```

```python
Y.head()
```

```python
###Feature Importance
from sklearn.ensemble import ExtraTreesRegressor
model=ExtraTreesRegressor()
model.fit(X,Y)
```

```python
#which features are important
print(model.feature_importances_)
```

```python
#plotting important features
p=pd.Series(model.feature_importances_,index=X.columns)
p.nlargest(5).plot(kind='barh')#plotting top 5 important features
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
##Hyperparamter tuning
n_estimators=[int(x) for x in np.linspace(start=100,stop=1400,num=14)]
print(n_estimators)
```

```python
n_estimators=[int(x) for x in np.linspace(start=100,stop=1400,num=14)]
max_features=['auto','sqrt']
max_depth=[int(x) for x in np.linspace(start=4,stop=32,num=8)]
min_samples_split=[2,5,10,15,32]
min_samples_leaf=[1,2,5,6,7,10]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
rf=RandomForestRegressor()
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10,
                               cv = 5,random_state=42, n_jobs =-1)
```

```python
rf_random.fit(X_train,Y_train)
```

```python
Y_pred=rf_random.predict(X_test)
Y_pred
```

```python
sns.distplot(Y_test-Y_pred)
```

```python
plt.scatter(Y_test,Y_pred)
```

```python
from sklearn.metrics import *
```

```python
print('MAE:', mean_absolute_error(Y_test,Y_pred))
print('MSE:', mean_squared_error(Y_test,Y_pred))
print('RMSE:',np.sqrt(mean_squared_error(Y_test,Y_pred)))
```

```python
import pickle
# open a file, where you ant to store the data
file = open('random_forest_regression_model.hd5', 'wb')

# dump information to that file
pickle.dump(rf_random, file)
```
